```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_data=pd.read_csv('/content/boston_house_pricesshv.csv')
```

```python
house_data.head()
```

```python
print(house_data)
```

```python
house_data.describe()
```

```python
corre=house_data.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corre,cbar=True,square = True,fmt ='.3f',annot=True,annot_kws={'size':8},cmap='Blues' )
```

```python
X=house_data.drop(['MEDV'],axis=1)
Y=house_data['MEDV']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=2)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
score_1 = metrics.r2_score(Y_train,training_data_prediction)
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print(score_1)
```

```python
print(score_2)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
score_3 = metrics.r2_score(Y_test,test_data_prediction)
score_4 = metrics.mean_absolute_error(Y_test,test_data_prediction)
```

```python
print(score_3,score_4)
```
