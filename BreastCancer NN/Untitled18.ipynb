```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
```

```python
data_frame = pd.DataFrame(breast_cancer_data.data,columns = breast_cancer_data.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_data.target
```

```python
data_frame.tail()
```

```python
data_frame.describe()
```

```python
data_frame.isnull().sum()
```

```python
data_frame['label'].value_counts()
```

```python
X = data_frame.drop(columns = 'label',axis = 1)
Y = data_frame['label']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2 ,random_state = 2)
```

```python
print(X_test.shape)
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
```

```python
trainning_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print(trainning_data_accuracy)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
        keras.layers.Flatten(input_shape = (30,)),
        keras.layers.Dense(20, activation = 'relu'),
        keras.layers.Dense(2,activation = 'sigmoid')

])
```

```python
model.compile(optimizer = 'adam',
              loss='sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
history = model.fit(X_train_std,Y_train,validation_split = 0.1, epochs = 10)
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy,loss)
```

```python
Y_train_pred = model.predict(X_train_std)
Y_test_pred = model.predict(X_test_std)
```

```python
print(Y_train_pred)
print(Y_test_pred)
```
